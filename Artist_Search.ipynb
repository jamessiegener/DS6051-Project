```python
def artist_id_search(artist_ist):
    '''Search Artists by IDs and put relevant metadata in dataframe'''
    
    results = sp.artists(artists = artist_list)

    artists = []

    # Print artist details
    for artist in results["artists"]:
        artists.append({'id': artist['id'], 'name': artist['name'], 'genres': artist['genres'], 'popularity': artist['popularity']})

    
    return pd.DataFrame(artists)

def artist_name_search(name):
    '''function to grab relevant data from query results and put it in data frame'''
    
    artist_name = name
    results = sp.search(q=artist_name, limit = 50, type="artist")

    artists = []

    # Print artist details
    for artist in results["artists"]["items"]:
        artists.append({'name': artist['name'], 'genres': artist['genres'], 'popularity': artist['popularity'], 'uri': artist['uri']})

    # example of where to find information in search result json    
    # print(f"Name: {artist['name']}, Followers: {artist['followers']['total']}, URL: {artist['external_urls']['spotify']}")
    
    return artists


def artist_name_search_loop(artist_list):
    '''Loop to get multiple artists from spotify,
        query gathers 50 results from each artist in artist_list'''
    artists = []
    
    # loop over artist list
    for char in artist_list:
        artist_name = char
        # make query
        results = sp.search(q=artist_name, limit = 50, type="artist")

        # Print artist details
        for artist in results["artists"]["items"]:
            artists.append({'name': artist['name'], 'genres': artist['genres'], 'popularity': artist['popularity'], 'uri': artist['uri']})

    return pd.DataFrame(artists)


def artist_name_search_loop_large(alphabet):
    '''loop to get multiple artists from spotify based on list of alphabetic characters,
        searches 676 combinations of letters with 50 outputs from each search'''
    
    artists = []
    
    # loop over artist list
    for char in alphabet:
        for char2 in alphabet:
            artist_name = char + char2
            # make query
            results = sp.search(q=artist_name, limit = 50, type="artist")

            # Print artist details
            for artist in results["artists"]["items"]:
                artists.append({'name': artist['name'], 'genres': artist['genres'], 'popularity': artist['popularity'], 'uri': artist['uri']})

    return pd.DataFrame(artists)
```
