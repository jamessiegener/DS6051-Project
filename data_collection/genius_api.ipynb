```python
import os
import requests
import json
import pandas as pd
import lyricsgenius
import dotenv
dotenv.load_dotenv()
```

Note: We worked with Genius API data in DS 6100 under lab assignment 4. I follow a similar process here.

```python
# Load in API keys
genius_id = os.getenv("GeniusID")
genius_secret = os.getenv("GeniusSecret")
genius_token = os.getenv("GeniusToken")
```

```python
# Get user-agent 
r = requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
headers = {'User-agent': useragent}
```

### Taylor Swift Data

```python
# Taylor Swift API call
params = {'access_token': genius_token,
         'q': 'Taylor Swift'}
root = 'https://api.genius.com'
endpoint = '/search'
r = requests.get(root+endpoint, headers=headers, params=params)
r
```

```python
taylor_json = json.loads(r.text)
taylor_json['response']['hits']
```

```python
# Load songs data into df
taylor_df = pd.json_normalize(json.loads(r.text), record_path=['response','hits'])
taylor_df
```

```python
# Get API path for "Down Bad"
downbad_path = taylor_json['response']['hits'][9]['result']['api_path']
downbad_path
```

```python
# Down Bad API call
params = {'access_token': genius_token}
root = 'https://api.genius.com'
r = requests.get(root+downbad_path, headers=headers, params=params)
r
```

```python
downbad_json = json.loads(r.text)
downbad_json['response']['song']
```

### Bob Dylan Data

```python
# Bob Dylan API call
params = {'access_token': genius_token,
         'q': 'Bob Dylan'}
root = 'https://api.genius.com'
endpoint = '/search'
r = requests.get(root+endpoint, headers=headers, params=params)
r
```

```python
dylan_json = json.loads(r.text)
dylan_path = dylan_json['response']['hits'][0]['result']['primary_artist']['api_path']
dylan_path
```

```python
# Sort Bob Dylan songs on popularity
endpoint = dylan_path + '/songs'
params = {'access_token': genius_token,
         'sort': 'popularity'}
r = requests.get(root+endpoint, headers=headers, params=params)
r
```

```python
dylan_json = json.loads(r.text)
dylan_json['response']['songs']
```

```python
# Load Bob Dylan song data into df
dylan_df = pd.json_normalize(json.loads(r.text), record_path=['response','songs'])
dylan_df.head()
```
